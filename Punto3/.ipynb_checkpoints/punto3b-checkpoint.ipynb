```python
import scikits.audiolab as audio
from scipy.io import wavfile
from scipy.fftpack import fft, fftfreq
import matplotlib.pylab as plt
import numpy as np
```

```python
sampling_rate, input_signal = wavfile.read('audio.wav')
print("sampling rate = {} Hz, length = {} samples, channels = {}".format(sampling_rate, *input_signal.shape))
print(input_signal[50000:50010])
plt.plot(input_signal)
```

```python
time_array = np.arange(0, len(input_signal)/float(sampling_rate), 1/float(sampling_rate))
```

```python
plt.plot(time_array[0:4000], input_signal[0:4000])
plt.xlabel("time(s)", fontsize=20)
plt.ylabel("Amplitude", fontsize=20)
#xlim(0.04,0.06)
```

```python
dt = 1 / (sampling_rate * len(input_signal)/4.)
n = len(input_signal)
from scipy.fftpack import fft, fftfreq
fft_x = fft(input_signal) / n # FFT Normalized
freq = fftfreq(n, dt) # Recuperamos las frecuencias
plot(freq,np.abs(fft_x))
print(freq)
```

```python
fft_x_shifted = np.fft.fftshift(fft_x)
freq_shifted = np.fft.fftshift(freq)
print(freq_shifted)
```

```python
plt.plot(freq_shifted, np.abs(fft_x_shifted))
plt.xlabel("Frecuencia (Hz)")
```

```python
figure(figsize(10,10))
half_n = np.ceil(n/2.0)
fft_x_half = (2.0) * fft_x[:half_n]
freq_half = freq[:half_n]
ampl = np.abs(fft_x_half)
plot(freq_half/1e9,ampl)
xlabel("Frecuencia (Hz)")
ylabel("Amplitud")
annotate('Primer armonico ~ 0.5',xy=(0.48,500), xycoords='data',textcoords='axes pixels',fontsize=10, arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
annotate('Segundo armonico ~ 0.9',xy=(0.9,300), xycoords='data',textcoords='axes points',fontsize=10, arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
annotate('Tercer armonico ~ 1.1',xy=(1.1,500), xycoords='data',textcoords='offset points',fontsize=10, arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
```
