#Hay calentamiento global?

Carguemos los paquetes necesarios

```python
%pylab inline
import pandas as pd
from scipy import interpolate
```

Carguemos los datos

```python
df = pd.read_csv('Calentamientoglobal.csv')
```

Convirtamos cada ciudad en un dataframe

```python
bog = df[df.ciudad == "Bogota"]
cal = df[df.ciudad == "Cali"]
buc = df[df.ciudad == "Bucaramanga"]
bar = df[df.ciudad == "Barranquilla"]
ipi = df[df.ciudad == "Ipiales"]
```

Eliminando columnas innecesarias

```python
del bog['anio']
del bog['mes']
del bog['ciudad']
del cal['anio']
del cal['mes']
del cal['ciudad']
del buc['anio']
del buc['mes']
del buc['ciudad']
del bar['anio']
del bar['mes']
del bar['ciudad']
del ipi['anio']
del ipi['mes']
del ipi['ciudad']
```

Convirtiendo de Pandas a listas para poder grafiar con matplotlib

```python
bogts = bog['fecha'].tolist()
bogtsint = linspace(0,len(bogts),len(bogts))
bogdat = bog['temperatura'].tolist()

calts = cal['fecha'].tolist()
caltsint = linspace(0,len(calts),len(calts))
caldat = cal['temperatura'].tolist()

bucts = buc['fecha'].tolist()
buctsint = linspace(0,len(bucts),len(bucts))
bucdat = buc['temperatura'].tolist()

barts = bar['fecha'].tolist()
bartsint = linspace(0,len(barts),len(barts))
bardat = bar['temperatura'].tolist()

ipits = ipi['fecha'].tolist()
ipitsint = linspace(0,len(ipits),len(ipits))
ipidat = ipi['temperatura'].tolist()

```

Una cosa que hay que tener en cuenta es que la interpolacion por splines suelen ser preferida sobre la interpolacion polinomica pues a medida que se aumenta el orden del polinomio este tiende a oscilar de manera "innecesaria" a continuacion se muestra dicho fenomeno.

```python
#bogota zoom
figure(figsize=(17,5))
f = interpolate.interp1d(bogtsint,bogdat)
f2 = interpolate.interp1d(bogtsint,bogdat,kind='quadratic')
tck = interpolate.splrep(bogtsint,bogdat, s=0)
xnew = np.arange(0,len(bogts),0.1)
f3 = interpolate.splev(xnew, tck, der=0)
plt.plot(bogtsint,bogdat, "ko",xnew,f(xnew),'-r',xnew,f2(xnew),'--g',xnew,f3,'c')
legend(['Datos', 'Lineal', 'Polinomio 4' ,'Splines 3'], loc='best')
xlim(20,30)
title('Cambios de temperatura a traves del tiempo en Bogota', fontsize=20)
```

Por otro lado la interpolacion lineal no es la óptima para este tipo de datos, pues los cambios de temperatura ocurren como procesos continuos. De manera que el mejor se debe escoger entre el spline y el polinomio (de grado correcto) siendo exagerados se puede ver a continuacion que el ajuste para splines es el mismo que el polinomico. Y por tanto no se puede determinar cual de los se ajusta mejor a los datos.

```python
#bogota zoom
figure(figsize=(17,5))
f2 = interpolate.interp1d(bogtsint,bogdat,kind='cubic')
tck = interpolate.splrep(bogtsint,bogdat, s=0)
xnew = np.arange(0,len(bogts),0.1)
f3 = interpolate.splev(xnew, tck, der=0)
plt.plot(bogtsint,bogdat, "ko",xnew,f2(xnew),'--r',xnew,f3,'y')
legend(['Datos', 'Polinomio 3' ,'Splines 3'], loc='best')
xlim(22,22.0001)
ylim(13.2,13.25)
title('Cambios de temperatura a traves del tiempo en Bogota', fontsize=20)
```

Examinemos los promedios de temperatura y su desviacion estandar para determinar si existe o no el calentamiento global 

```python
mean(bogdat), std(bogdat)
```

```python
mean(caldat), std(caldat)
```

```python
mean(bucdat), std(bucdat)
```

```python
mean(bardat), std(bardat)
```

```python
mean(ipidat), std(ipidat)
```

En todos los casos el promedio es un valor perfectamente aceptable entre la serie de temperaturas. La desviacion estandar es muy pequéña en todos los casos de manera que se indica que se mantienen de alguna manera "anclados" al rededor del promedio. Y las series no muestran comportamientoes ascendentes (ni descendentes), de manera que se puede concluir, en base a los datos analizados que no existe calentamiento global.
